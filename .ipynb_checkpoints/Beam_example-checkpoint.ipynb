```python
def fem_solution(loadvec):
    #Element stiffnes matrix
    k = np.matrix([[12, 6, -12, 6], [6, 4, -6, 2], [-12, -6, 12, -6], [6, 2, -6, 4]])*10/12
    
    #Global stiffness matrix
    n_dofs = n_nodes*2
    K = np.zeros([n_dofs, n_dofs])
    for i in range(n_elements):
        K[2*i:2*i+4,2*i:2*i+4] += k
    
    #Loadvector with bc
    F = np.zeros(n_dofs)
    F[0::2] = loadvec
    
    #Account for homogenous BC
    F = F[2:-2]
    K = K[2:-2,2:-2]
    
    #Solve matrix equation
    d = np.matmul(inv(K),F)
    
    #Vertical displacement vector
    dz = d[0:-1:2]
    dz = np.insert(dz,0,0)
    dz = np.append(dz,0)
    
    return dz
```

```python
#Create training data
n_training = 200

X_train = np.zeros([n_nodes,n_training])
Y_train = np.zeros([n_nodes,n_training])

for i in range(n_training):
    #Generate random loadvector
    loadvec = np.random.random(n_nodes)
    X_train[:,i] = loadvec
    
    dz = fem_solution(loadvec)
    Y_train[:,i] = dz
    
```

```python
#Neural Network

#f = pd.read_csv('3D_beam_frequencies.csv', header=None)
#u = pd.read_csv('3D_beam_maximum_displacements.csv', header=None)


x_train = X_train[:,0:150]
y_train = Y_train[:,0:150]
x_test = X_train[:,150:]
y_test = Y_train[:,150:]

def baseline_model():
    model = Sequential()
    height = 200
    model.add(Dense(height, input_dim=n_nodes, activation='tanh', kernel_initializer='normal'))
    model.add(Dense(height, input_dim=height, activation='tanh', kernel_initializer='normal'))
    model.add(Dense(height, input_dim=height, activation='tanh', kernel_initializer='normal'))
    model.add(Dense(n_nodes, input_dim=height, activation='linear', kernel_initializer='normal'))
    
    model.compile(loss='mse', optimizer="adam")
    return model



# Training the model
model = baseline_model()
model.fit(np.transpose(x_train), np.transpose(y_train), epochs=500, verbose = 0)
```

```python
plt.figure(figsize=(16,10))

plt.plot(x, y_test)

plt.show()
```
